```python
new_tiger=tiger.reshape((tiger.shape[0]* tiger.shape[1]),tiger.shape[2])
new_dog=dog.reshape((dog.shape[0]* dog.shape[1]),dog.shape[2])
new_tiger.shape,new_dog.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
labels=kmeans.fit_predict(new_dog)
```

```python
labels=labels.reshape(dog.shape[0],dog.shape[1])
```

```python
plt.imshow(labels)
```

```python
centers = [centers.astype(int) for centers in kmeans.cluster_centers_]
blob = np.ones((100,100,3))
blob1 = blob * centers[0]
#blob2=blob * centers[1]
#blob3= blob * centers[2]
```
